```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.preprocessing import LabelEncoder

import seaborn as sns
sns.set()

import warnings
warnings.filterwarnings("ignore")
```

### Data Preparation and Preprocessing

```python
#The path to the csv file
file_path = "../data/Housing.csv"
#Load the dataset
dataset = pd.read_csv(file_path)
#Get first 5 elements
dataset.head()
```

```python
#Check for missing values
dataset.isna().sum()
```

```python
#Get dataset shape
dataset.shape
```

```python
#Get dataset information
dataset.info()
```

```python
#Change nominal categorical columns to numerical values
#Columns to map
columns_to_transform = ["mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea"]
#Mapping dictionary
dict_to_map ={"yes":1, "no":0}
#Map
dataset[columns_to_transform] = dataset[columns_to_transform].replace(dict_to_map)
```

```python
#Get first 5 elements
dataset.head()
```

```python
#Create label encoder object
label_encoder = LabelEncoder()
#Apply label encoding to the furnishingstatus column
dataset["furnishingstatus"] = label_encoder.fit_transform(dataset["furnishingstatus"])
```

```python
#Get first 5 elements
dataset.head()
```
